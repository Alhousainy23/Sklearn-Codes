```python
#Using For Loop For Selected Features 
#Another Example Using Selected Features 
from sklearn.datasets import load_breast_cancer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectPercentile,f_classif,chi2
data2 = load_breast_cancer()
x2_data=data2.data
y2_target = data2.target
for j in range (10,100,5):
    Sel =SelectPercentile(score_func=chi2,percentile=j)
    Sel.fit(x2_data,y2_target)
    selected_features=Sel.transform(x2_data)
    SFeatures=Sel.get_support()
    #print('\t\t The Selected Features Is Equal \n',SFeatures)
    #print('***'*30)
    #Splitting Into Data To Training & Test Data 
    x2_train,x2_test,y2_train,y2_test=train_test_split(selected_features,y2_target,test_size=0.2)
    LogReg_Model4=LogisticRegression()
    LogReg_Model4.fit(x2_train,y2_train)
    result4=LogReg_Model4.predict(x2_test)
    print('For Precentile is Equal ==> ',j)
    print('The Accuracy Score Is ==> ',accuracy_score(y2_test,result4))
    #print('**'*20)
    #conf=confusion_matrix(y2_test,result4)
    #print('\t\t The Confusion Matrix Is \n',conf)
    print('=======================================================================================================')
```

```python
#Classfication Using Select From Model 
from sklearn.datasets import load_breast_cancer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectFromModel
data=load_breast_cancer()
x_data=data.data    
y_target = data.target        
sel =SelectFromModel(RandomForestClassifier(n_estimators=20))
sel.fit(x_data,y_target)
selected_features1=sel.transform(x_data)
SFeatures1=sel.get_support()
print('Selected Features Are \n',SFeatures1)
print('==='*30)
x_train,x_test,y_train,y_test=train_test_split(selected_features1,y_target,test_size=0.2)
LogReg_Model=LogisticRegression()
LogReg_Model.fit(x_train,y_train)
result=LogReg_Model.predict(x_test)
print('Accuracy Score Is ==> ',accuracy_score(y_test,result))
print('=='*30)
print('\t\t The Confusion Matrix Are \n',confusion_matrix(y_test,result))
print('**'*40)
```
